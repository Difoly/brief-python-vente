```python
# importation des bibliotheques
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# importation du fichier des vente de l'entreprise 
ventes = pd.read_csv("ventes-data.csv")
ventes
```

```python
# verification des valeurs aberrantes
ventes.describe()
```

```python
# verification des valeurs manquantes
ventes.isnull().sum()
```

```python
# calcul des revenus par mois et clssification par ordre decroissant 
revenu_mois = ventes.groupby("Month")[["Revenue"]].sum().sort_values("Revenue", ascending = False)
revenu_mois
```

```python
# les 5 produits les plus vendu par unite de produits chaque mois
top = ventes.groupby(["Month","Product"])[["Units Sold"]].sum()
top1 = top.sort_values(by =["Month", "Units Sold"], ascending = [True, False])
top5 = top1.groupby("Month").head(5)
top5
```

```python
# les 5 produits les plus vendu par revenu  chaque mois
top = ventes.groupby(["Month","Product"])[["Revenue"]].sum()
top1 = top.sort_values(by =["Month", "Revenue"], ascending = [True, False])
top5 = top1.groupby("Month").head(5)
top5
```

```python
# les mois avec les ventes les plus elees et les revenus les plus basses
eleve = revenu_mois.head(5)
bas = revenu_mois.tail(5)
print(f"les mois avec les ventes les plus elevees \n{eleve} \n les mois avec les revenus les plus bas \n{bas}")
```

```python
x = ventes['Month']
y = ventes['Revenue']
plt.barh(x, y)
plt.title("Innova Electronic: Revenu Par Mois")
plt.xlabel("Revenue")
plt.ylabel("Mois")
plt.show()
```

```python
# graphique du nombre des unites vendu par produits
x = ventes['Product']
y = ventes['Revenue']
z = ventes["Units Sold"]
plt.figure(figsize=(8, 8))
plt.barh(x, z)
plt.title("Innova Electronic: unite vendu par  produit")
plt.xlabel("revenue")
plt.ylabel("prodiuts")
plt.show()
```

```python
# graphyque revenu par produit 
x = ventes['Product']
y = ventes['Revenue']
z = ventes["Units Sold"]
plt.figure(figsize=(8, 8))
plt.barh(x, y)
plt.title("Innova Electronic: revenu par produit")
plt.xlabel("revenu")
plt.ylabel("prodiuts")
plt.show()
```

```python
ventes['taux_Revenu'] = round(ventes["Revenue"]*100/ventes["Revenue"].sum(), 2)
ventes['taux_units_sold'] = round(ventes["Units Sold"]*100/ventes["Units Sold"].sum(), 2)
ventes
```

```python
ventes.describe()
```

```python
x = ventes['Product']
y = ventes['taux_Revenu']
z = ventes["Units Sold"]
plt.figure(figsize=(8, 8))
plt.barh(x, y)
plt.title("Innova Electronic: taux de revenu")
plt.xlabel("taux")
plt.ylabel("prodiuts")
plt.show()
```

```python
# graphique du taux d  
x = ventes['Product']
y = ventes['taux_units_sold']
plt.figure(figsize=(8, 8))
plt.barh(x, y)
plt.title("Innova Electronic: taux d'unite vendu")
plt.xlabel("taux")
plt.ylabel("prodiuts")
plt.show()
```

```python
# histogramme des revenus
x = ventes['Month']
y = ventes['Revenue']
z = ventes["Units Sold"]
plt.hist(y,color = 'green', rwidth = 0.8)
plt.title("Innova Electronic: histogramme des revenus")
plt.xlabel("revenu")
plt.ylabel("effectifs")
plt.show()
```
